```python
from megengine import tensor
import matplotlib.pyplot as plt
import numpy as np
import megengine.module as M
from megengine.data.transform import Compose,Pad,RandomResizedCrop


import cv2

img=cv2.imread("D:\HZY\Porio\ComputerCode\Jupyter\images\img\cat3.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
# print(img)
plt.imshow(img)
plt.show()
img=img.transpose([2,0,1])[np.newaxis,:].astype(np.float32)
```

```python
from PIL import Image
m2 = M.Conv2d(3,3,kernel_size=20)
out = m2(tensor(img)).numpy()
out= out[0].transpose([1,2,0]).astype(np.uint8)
plt.imshow(out)
im = Image.fromarray(out)
im.save("D:\HZY\Porio\ComputerCode\Jupyter\images\img\out.jpg")
# plt.savefig("D:\\HZY\\学业\\课程\\2022春季学期\\数字图像处理\\课设\\处理图片\\cat3_output.jpg")
plt.show()
```
